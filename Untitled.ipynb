```r
library(tidyverse)
library(repr)
library(tidymodels)
options(repr.matrix.max.rows = 6)
```

```r
pulsar <- read_csv("pulsar_data/HTRU_2.csv", 
                   col_names = c("M_IP","SD_IP","EK_IP","SK_IP","M_DM_SNR","SD_DM_SNR","EK_DM_SNR","SK_DM_SNR","CLASS")) |>
                mutate(CLASS = as_factor(CLASS))
pulsar
```

```r
pulsar_summarize <- pulsar|>
            group_by(CLASS)|>
            summarize(across(M_IP:SK_DM_SNR, mean))
pulsar_summarize
```

```r
pulsar_na <- pulsar|>
             summarise(across(everything(), ~ sum(is.na(.))))
pulsar_na
```

```r
type_labels = c("0" = "Non-pulsar", "1" = "Pulsar")
pulsar_plot_mip <- pulsar |>
                ggplot(aes(x = M_IP, fill = CLASS)) +
                geom_histogram() +
                facet_grid(rows = vars(CLASS),
                          labeller = labeller(CLASS = type_labels)) +
                labs(x = "Mean Integrated Profile", fill = "Type") +
                scale_fill_manual(labels = c("Non-pulsar", "Pulsar"),
                                 values = c("orange2", "steelblue")) +
                scale_y_log10() +
                theme(text = element_text(size = 20))
pulsar_plot_mip
```
